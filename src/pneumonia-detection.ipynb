```python
# Lectura de la imagen

img = pydicom.dcmread("../input/rsna-pneumonia-detection-challenge/stage_2_train_images/0004cfab-14fd-4e49-80ba-63a80b6bddd6.dcm").pixel_array
plt.imshow(img)
```

```python
# Data augmentation

# img = np.fliplr(img)
# plt.imshow(img)

img_gen = ImageDataGenerator(samplewise_center=False, 
                             samplewise_std_normalization=False, 
                             horizontal_flip = True, 
                             vertical_flip = False, 
                             height_shift_range = 0.05, 
                             width_shift_range = 0.02, 
                             rotation_range = 3, 
                             shear_range = 0.01,
                             fill_mode = 'nearest',
                             zoom_range = 0.05)
# img_gen = ImageDataGenerator(width_shift_range=[-200,200])

img = img.reshape(1,1024,1024,1) # (1, dim1, dim2, canal)

it = img_gen.flow(img, batch_size = 1)

# generate samples and plot
for i in range(9):
    # define subplot
    plt.subplot(330 + 1 + i)
    # generate batch of images
    batch = it.next()
    # convert to unsigned integers for viewing
    image = batch[0].astype('uint8')
    # plot raw pixel data
    plt.imshow(image)
# show the figure
plt.show()
```

```python
# Normalización por imagen

img_min = img.min()
img_max = img.max()

(img - img_min) / (img_max - img_min)
```
